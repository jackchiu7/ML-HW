```python
import matplotlib.pyplot as plt
from math import e
```

```python
year = range(1930,2020,10)
gdp = [1.015, 1.33, 2.29, 3.26, 4.951, 6.759, 9.366, 13.131, 15.599]
```

```python
plt.plot(year, gdp, 'ro', year, gdp, 'r--')
plt.ylabel('Trillions of $')
plt.xlabel('Year')
plt.title('Real US GDP (in trillions)')
```

Based on the plotted data, the GDP and year could have be in the form of $gdp = (year)^2$. It is quite apparent that the relationship can not be linear. Taking $1930 = 1$, $1940 = 2$, and so forth, if the relationship was linear, $f(2010) \approx 9$ since $f(1930) \approx 1$. However, $f(2010) \approx 16$. We take the  possibility that the relationship is in the form of $f(x) = x^2$.

```python
gdpApprox = []
for i in range(1,10):
    gdpApprox.append(i**2/5.5+.8)
plt.plot(year, gdp, 'ro', year, gdp, 'r--')
plt.ylabel('Trillions of $')
plt.xlabel('Year')
plt.title('Real US GDP (in trillions)')
plt.plot(year, gdpApprox, 'b-')
```

Playing around with a scalar multiplier and shifting the approximated graph up or down yields the above equation $gdp = (\frac{year - 1920}{10})^2 + 0.8$.
